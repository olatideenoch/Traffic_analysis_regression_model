# Simple Linear Regression

## Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## Importing the dataset
- Identifying the Target and Features

```python
traffic_df = pd.read_csv("Traffic_data.csv")
X = traffic_df.iloc[:, 1:2].values
y = traffic_df.iloc[:, -1:].values
print(X)
print(y)
```

## Splitting the dataset into the Training and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

## Training and Testing the Simple Linear Regression model on the Training set

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
pred = regressor.predict(X_train)
print(pred)
```

## Visualizing the Training set results

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, pred, color='blue')
plt.title('Traffic Volume analysis over time')
plt.ylabel('Traffic Volume')
plt.xlabel('Datetime')
plt.show()
```
